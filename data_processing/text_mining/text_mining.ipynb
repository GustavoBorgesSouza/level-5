## Tokenização

Pocessar partes de uma frase em outras menores (tokens)

```python
import nltk
from nltk.corpus import machado

stopwords = nltk.corpus.stopwords.words('portuguese')
print(stopwords)

sent_tokenizer = nltk.data.load('tokenizers/punkt/portuguese.pickle')

raw_text = machado.raw('romance/marm05.txt')

sentences = sent_tokenizer.tokenize(raw_text)

for sent in sentences[10:15]:
    print("\n<<", sent, '>>')
```

## Análise léxica

Extração de lemmas/radicais

```python
import nltk
from nltk.stem import *
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer

stemmer = PorterStemmer()


plurals = ['caresses', 'flies', 'dies', 'mules', 'denied','died', 'agreed', 'owned', 'humbled', 'sized','meeting', 'stating', 'siezing', 'itemization','sensational', 'traditional', 'reference', 'colonizer','plotted']

singles = [stemmer.stem(plural) for plural in plurals]

print(singles)
print(SnowballStemmer.languages)

stemmer2 = SnowballStemmer("portuguese", ignore_stopwords=True)
print(stemmer2.stem('comer'))
print(stemmer2.stem('feliz'))
print(stemmer2.stem('livre'))
```

##  Análise sintática

construções de árvores sintáticas para entender como a estrutura de uma sentença é construida

```python
import nltk
from nltk.tree import *
from nltk.corpus import wordnet as wn

dp1 = Tree('dp', [Tree('d', ['o']), Tree('np', ['cachorro'])])
dp2 = Tree('dp', [Tree('d', ['o']), Tree('np', ['gato'])])
vp = Tree('vp', [Tree('v', ['perseguiu']), dp2])
tree = Tree('s', [dp1, vp])

print(tree)

dp1.label(), dp2.label(), vp.label(), tree.label()

print(tree.treepositions())
```

```python
print(tree.pformat_latex_qtree())
print(tree.pretty_print())
print(tree.pretty_print(unicodelines=True, nodedist=4))
```

## Análise semântica

Da o contexto, sentido as palavras/frases/orações

```python
import nltk
from nltk.corpus import floresta

print(floresta.words()[:10])

print(floresta.tagged_words()[:10])

def simplify_tag(t):
    if "+" in t:
        return t[t.index("+")+1:]
    else:
        return t

t_words = floresta.tagged_words()

t_words = [(w.lower(), simplify_tag(t)) for (w, t) in t_words]


print(t_words[:10])
```

## Análise pragmática

extriair e compreender o que está sendo dito no texto

```python
import nltk
import random

def gender_features(name):
    name = name.lower()
    return {
        'ultimo_caractere': name[-1],
        'ultimos_dois': name[-2:],
        'ultimo_tres': name[-3:],
        'primeiro_caractere': name[0],
        'segundo_caractere': name[:1],
    }

f_names = nltk.corpus.names.words('female.txt')
m_names = nltk.corpus.names.words('male.txt')

all_names = [(i, 'masculino') for i in m_names] + [(i, 'feminino') for i in f_names]
random.shuffle(all_names)

divide_lenght = int(len(all_names) / 2)

test_set, train_set = all_names[divide_lenght:], all_names[:divide_lenght]

test_set_feat = [(gender_features(n), g) for n, g in test_set]
train_set_feat = [(gender_features(n), g) for n, g in train_set]

classifier = nltk.NaiveBayesClassifier.train(train_set_feat)

classifier.classify_many([gender_features('Woody'), gender_features('Sabrina'), gender_features('Neo'), gender_features('Tito')])
```

```python
classifier.show_most_informative_features(10)
```

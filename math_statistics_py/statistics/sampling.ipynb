```python
import numpy as np
import pandas as pd
import random as rd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r'math_statistics_py/statistics/dados_renda_municipios.csv', sep=',', decimal= '.', encoding = 'utf-8')

data
```

```python
uf = 'São Paulo'
cities_data = data[data['UF'] == uf].reset_index(drop=True)

cities_data['classe_renda'] = pd.qcut(cities_data['RDPC'], 4, labels=['D', 'C', 'B', 'A'])
cities_data
```

```python
dados_piloto = cities_data.agg(mean_RDPC = pd.NamedAgg('RDPC', 'mean'),
dp_RDPC = pd.NamedAgg('RDPC', 'std'),
N = pd.NamedAgg('RDPC', 'count'))

dados_piloto
```

```python
# dados piloto por estrato
dados_piloto_classe = cities_data.groupby('classe_renda').agg(media_RDPC = pd.NamedAgg('RDPC', 'mean'),
dp_RDPC = pd.NamedAgg('RDPC', 'std'),
N = pd.NamedAgg('RDPC', 'count')) .reset_index()

dados_piloto_classe['percent'] = dados_piloto_classe['N'] / sum(dados_piloto_classe['N'])
dados_piloto_classe
```

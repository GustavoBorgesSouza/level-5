```python
import numpy as np
import pandas as pd

df_countries = pd.read_csv('math_statistics_py/statistics/dados_1997_2011_paises_csv.csv', sep=';', encoding='latin1', decimal=',')

df_countries
```

```python
df_countries.groupby('pais').agg(num = pd.NamedAgg('pais', 'count')).reset_index()
```

```python
df_countries.dtypes
```

## Análises e funções python

| análises | funções |
|----------|---------|
| Mínimo, Máximo, Média | min(), max(), mean() |
| Moda | mode() |
| Mediana | median() |
| Quartis, Percentis | quantile() |
| Desvio medio absoluto | mad() |
| Variância, Desvio Padrão | var(), std() |
| Contagem | count() |

```python
df_countries.groupby('pais')['idh'].apply(lambda x: x.mode().iloc[0]).to_frame().reset_index()
```

```python
df_countries.groupby('pais')['idh'].apply(lambda x: x.quantile([0.05,0.25,0.5,0.75,0.95])).to_frame().reset_index().rename(columns={'level_1': 'percentil'})
```

```python
df_countries.groupby('pais').agg(desvio_medio_abs_idh = pd.NamedAgg( column='idh', aggfunc=lambda x: x.mad())).reset_index()
```

```python
df_countries.groupby('pais').agg(variancia_idh = pd.NamedAgg('idh', 'var'),
dp_idh = pd.NamedAgg('idh', 'std')).reset_index()
```

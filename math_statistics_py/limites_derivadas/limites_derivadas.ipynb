# Aplicações de limites e derivadas

## Libs 

```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
```

## Limites

```python
x = smp.Symbol('x')
# apply var as math symbol
x
```

```python
funcao = ((x**2 + x-2) / (x -1))

funcao
```

```python
# Limite da função quando variavel x tende a 1 
smp.limit(funcao, x, 1) 
```

```python
smp.plot(funcao)
```

```python
beta0 = smp.Symbol('beta0')
beta1 = smp.Symbol('beta1')

funcao_beta = (beta0 + beta1*x)

funcao_beta
```

```python
smp.limit(funcao_beta, x, 0)
```

```python
# limite tendendo ao infinito
smp.limit(funcao_beta, x, smp.oo)
```

```python
# analise limites
x = smp.Symbol('x')
funcao_reg_logs = 1 / (1+smp.exp(-x))

funcao_reg_logs
```

```python
# limite tendendo a zero
smp.limit(funcao_reg_logs, x, 0)
```

```python
# limite tendendo ao infinito 
smp.limit(funcao_reg_logs, x, smp.oo)
```

```python
# limite tendendo ao infinito negativo
smp.limit(funcao_reg_logs, x, -smp.oo)
```

```python
smp.plot(funcao_reg_logs)
```

## Derivadas

### Análise da regra da constante

```python
smp.diff(3)
```

### Análise da regra da potencia 

```python
x = smp.Symbol('x')
f = x**2

smp.diff(f)
```

```python
derivada = smp.diff(f)

original = smp.lambdify(x, f)
derivada = smp.lambdify(x, derivada)

# original chart
y = [original(i) for i in range(-100, 100)]
plt.plot(y)

# diff chart
y_d = [derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```

### Análise da regra da multiplicação

```python
f = (x**2 + 3) * (2*x +1)
f_d = smp.diff(f)

smp.simplify(f_d)
```

```python
original = smp.lambdify(x, f)
diff = smp.lambdify(x, f_d)

y = [original(i) for i in range(-100, 100)]
plt.plot(y)

y_d = [diff(i) for i in range(-100, 100)]

plt.plot(y_d)
```

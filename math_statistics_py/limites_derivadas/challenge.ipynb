```python
import sympy as smp
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = smp.Symbol('p')
funcao_lucro = (2*p**2 + 50.4 *p) - 10

funcao_lucro
```

```python
smp.plot(funcao_lucro)
```

```python
f_d = smp.diff(funcao_lucro)

f_d
```

```python
original = smp.lambdify(p, funcao_lucro)
diff = smp.lambdify(p, f_d)

y = [original(i) for i in range(-100,100)]
y_d = [diff(i) for i in range(-100,100)]

plt.plot(y)
plt.plot(y_d)
```
